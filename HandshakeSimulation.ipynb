```python
"""PROBLEM STATEMENT: Math 170E [Optional**] Write a computer code that simulates the hand shaking problem. That is, each
step pick a random pair of students and record that they have shaken hands. The simulation stops
when every single pair of students has shaken hands. Make a plot showing how many steps are
needed for class-sizes of n = 2, n = 4, n = 10, n = 20. (Even harder – run each simulation multiple
times and try to figure out the average number of hand shakes needed). From the graph try to test
the hypothesis that the number of hand shakes needed increases exponentially with n."""
```

```python
import matplotlib.pyplot as plt
import random
from matplotlib.pyplot import figure
```

```python
import statistics
def handshakesimulation(n,trials):
    totalposshake = (n*(n-1))/2
    #print(totalposshake)
    resultlist = []
    for x in range (0,trials):
        studentlist = list(range(0,n))
       #print(studentlist)
        pairhash = {}
        z = 0 
        while len(pairhash) < totalposshake:
            z = z + 1
            random_index1 = random.randint(0,len(studentlist)-1)
            random_index2 = random.randint(0,len(studentlist)-1)
            while random_index1 == random_index2: random_index2 = random.randint(0,len(studentlist)-1) 
            #print(str(random_index1) + " " + str(random_index2))
            keys = [random_index1, random_index2]
            keys.sort()
            dictkey = str(keys[0]) + str(keys[1])
            #print(dictkey)
            try:
                pairhash[dictkey] = (str(keys[0]),str(keys[1]))
            except:
                continue 
        #print(pairhash)
        resultlist.append(z)
    #print(resultlist)
    meanresult = statistics.mean(resultlist)
    #print(meanresult)
    f = plt.figure()
    f.set_figwidth(6)
    f.set_figheight(2)
    plt.title("Handshake Simulation of Class Size " + str(n) + " - 300 Trials")
    plt.boxplot(resultlist, vert = False)
    plt.show
    return meanresult
handshakesimulation(2,300)
handshakesimulation(4,300)
handshakesimulation(10,300)
handshakesimulation(20,300)
```

```python
#using only 50 trials so laptop doesn't explode
#would like to use histogram instead of a bar but gives me error
#class size range 0-45
x = []
y = []
for f in range(1,45):
    x.append(f)
    y.append(handshakesimulation(f,50))
print(x)
print(y)
plt.title("Handshake Simulation for Increasing Class Size - 50 Trials")
plt.bar(x,y)
plt.show
```
